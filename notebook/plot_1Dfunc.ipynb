```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w = [1,4,16]
t = np.arange(0,12,0.01)
n=1
plt.figure(figsize=(12,8))
for w0 in w:
  y = np.sin(w0*(t-np.pi)) 
  plt.subplot(len(w),1,n)
  plt.plot(t,y)
  plt.title(f'w: {w0}')
  plt.xlabel('t')
  plt.ylabel('f(t)')
  plt.tight_layout()
  plt.grid('on')
  n=n+1
plt.show()
```

```python
t = np.arange(-12,12,0.01)
t_ = np.arange(-24,24,0.01)
w0 = 1
f = np.cos(w0*t)
#g = np.ones((len(t),))
g = np.sin(w0*t)
h = np.convolve(f,g)
plt.figure(figsize=(12,8))

plt.plot(t_[:-1],h)
plt.title(f'convolution of f and g')
plt.xlabel('t')
plt.tight_layout()
plt.grid('on')
n=n+1
```
