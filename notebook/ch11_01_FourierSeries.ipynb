# A Basic example of Fourier series
- 주기함수는 서로 다른 주기를 가지는 코사인과 사인의 linear combination으로 표현이 가능합니다.
- Periodic rectangular wave를 통해 이를 확인해 보는 예제입니다.

```python
import numpy as np
from matplotlib import pyplot as plt
```

- 1000개의 숫자를 -2pi와 2pi사이에서 생성하여 x라 합니다.


```python
x = np.linspace(-2*np.pi,2*np.pi,1000)
```

- cos(nx)과 sin(nx)를 함수로 정의합니다.

```python
def cos(x,n):
  return np.cos(n*x)
def sin(x,n):
  return np.sin(n*x)
```

```python
plt.figure(dpi=80)
n = 8
plt.grid()
plt.plot(x,cos(x,n))
plt.title(f'cos({n}x)')
plt.axis([-2*np.pi,2*np.pi,-1.1,1.1])
plt.show()
```

## Periodic Rectangular Wave (11.1. Example 1)
- periodic rectangular wave의 Fourier series는 다음과 같다.
- a_0 = 0
- a_n = 0
- b_n = (2/(n*pi)) * ( 1 - cos(n*pi) )


```python
plt.figure(dpi=80)
plt.grid()
plt.plot(x,sin(x,1))
plt.show()
```

```python
plt.figure(dpi=80)
plt.grid()
plt.plot(x,sin(x,1) + (1/3)*sin(x,3))
plt.show()
```

```python
plt.figure(dpi=80)
plt.grid()
plt.plot(x,sin(x,1) + (1/3)*sin(x,3) + (1/5)*sin(x,5))
plt.show()
```

- n을 1부터 조금씩 늘려가면서 n이 커짐에 따라 어떤 함수로 수렴하는지 확인합니다.


```python
n = 500

a0 = 0
fx = a0*np.ones(np.shape(x))
for n in range(1,n+1):
  an = 0
  bn = (2./(n*np.pi))*(1-np.cos(n*np.pi))
  fx += an*cos(x,n)+bn*sin(x,n)

plt.figure(dpi=80)
plt.grid()
plt.plot(x,fx)
plt.show()
```
