```python
import numpy as np
from matplotlib import pyplot as plt
```

## 1. Direction Field of y
- y' = f(x,y)가 주어졌을 때 x,y에 대한 y'을 그래프로 나타낼 수 있다.
- 그래프의 결과를 관찰함으로써 solution curves를 추정해 볼 수 있다.

```python
# ODE y' = f(y, x) = y + x 가 주어졌다고 하자. 
def f(x,y):
    return y+x
```

```python
x = np.arange(0,3,0.2)
y = np.arange(0,3,0.2)
```

```python
plt.figure(dpi=150)
plt.grid()
for j in x:
    for k in y:
        slope = f(j,k)
        plt.quiver(j,k,1,slope,color='r',width=0.0035)
plt.title('Direction Field of y')
plt.xlabel('x')
plt.ylabel('y')
plt.axis([min(x)-.5,max(x)+.5,min(y)-.5,max(y)+.5])
plt.show()
```

## 2. Euler's method
- Euler's method는 ODE의 solution을 수치적인 방법으로 근사한다.
- 특정한 x0에 대한 y(x0) = y0 으로 주어져 있을 때, y(x0+h)의 값을 y'(x0)와 y(x0)의 값으로 부터 근사한다.
- 근사한 y(x0+h)는 h가 작을 수록 근사한 값과 실제 y(x0+h)값 사이의 차이가 작아질 수 있다.

```python
# ODE y' = f(x, y) = y + x 가 주어진 경우를 생각해 보자.
# 또한, initial condition으로 y(0) = 0 가 함께 주어졌다고 하자.
def f(x,y):
    return y+x
```

```python
x0 = 0
y0 = 0
x_last = 1
```

```python
x_ = np.array([])
y_ = np.array([])
x_ = np.append(x_,x0)
y_ = np.append(y_,y0)
plt.grid()
plt.scatter(x_,y_)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
h = 0.2
x_ = np.array([])
y_ = np.array([])
x_ = np.append(x_,x0)
y_ = np.append(y_,y0)
while x_[-1]<x_last:
    y_=np.append(y_, y_[-1]+f(x_[-1],y_[-1])*h)
    x_=np.append(x_, x_[-1]+h)
plt.plot(x_,y_)
plt.scatter(x_,y_)
plt.xlabel('x')
plt.ylabel('y_approximated')
plt.grid('on')
plt.show()

x_1 = x_
y_1 = y_
```

```python
h = 0.02
x_ = np.array([])
y_ = np.array([])
x_ = np.append(x_,x0)
y_ = np.append(y_,y0)
while x_[-1]<x_last:
    y_=np.append(y_, y_[-1]+f(x_[-1],y_[-1])*h)
    x_=np.append(x_, x_[-1]+h)
plt.plot(x_,y_)
plt.scatter(x_,y_)
plt.xlabel('x')
plt.ylabel('y_approximated')
plt.grid('on')
plt.show()

x_2 = x_
y_2 = y_
```

```python
plt.plot(x_1,y_1)
plt.plot(x_2,y_2)
plt.legend(['h=0.2','h=0.02'])
plt.xlabel('x')
plt.ylabel('y_approximated')
plt.grid('on')
plt.show()
```
