```python
import numpy as np
from matplotlib import pyplot as plt
```

## 1. Direction Field of y
- y' = f(x,y)가 주어졌을 때 x,y에 대한 y'을 그래프로 나타낼 수 있다. 
- 그래프의 결과를 관찰함으로써 solution curves를 추정해 볼 수 있다.

```python
def f(x,y):
    return y+x
```

```python
x = np.linspace(-4,4,20)
y = np.linspace(-4,4,20)
```

```python
plt.figure(dpi=100)
plt.grid()
for j in x:
    for k in y:
        slope = f(j,k)
        plt.quiver(j,k,1,slope,color='b',width=0.0035)
plt.title('Direction Field of y')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

## 2. Euler's method
- Euler's method는 ODE의 solution을 수치적인 방법으로 근사한다.
- 특정한 x0에 대한 y(x0) = y0 으로 주어져 있을 때, y(x0+h)의 값을 y'(x0)와 y(x0)의 값으로 부터 근사한다.
- 근사한 y(x0+h)는 h가 작을 수록 근사한 값과 실제 y(x0+h)값 사이의 차이가 작아질 수 있다.

```python
def f(x,y):
    return np.cos(x)

def sol(x):
    return np.sin(x)
```

```python
x = np.linspace(0,6.28,20)
y = np.linspace(0,6.28,20)
y_sol = sol(x)
```

```python
plt.figure(dpi=100)
plt.grid()
plt.plot(x,f(x,y))
plt.plot(x,y_sol)
plt.legend(["y'","y"])
plt.xlabel('x')
plt.show()
```

```python
x_ = np.linspace(0,6.28,20)
h = x_[2]-x_[1]
y_est = np.zeros(np.shape(x_))
y_est[0] = 0
for n in range(1,len(x_)):
    y_est[n] = y_est[n-1] + f(x_[n-1],y_est[n-1])* (x_[n]-x_[n-1])
```

```python
plt.figure(dpi=100)
plt.grid()
plt.plot(x_,sol(x_))
plt.plot(x_,y_est)
plt.legend(['solution','euler method'])
plt.xlabel('x')
plt.show()
```
