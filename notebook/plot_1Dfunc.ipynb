```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w = [1,4,16]
t = np.arange(0,12,0.01)
n=1
plt.figure(figsize=(12,8))
for w0 in w:
  y = np.sin(w0*t)
  plt.subplot(3,1,n)
  plt.plot(t,y)
  plt.title(f'w: {w0}')
  plt.xlabel('t')
  plt.ylabel('f(t)')
  plt.tight_layout()
  plt.grid('on')
  n=n+1
plt.show()
```
