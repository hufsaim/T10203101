# A Basic example of Fourier series

- 주기함수는 서로 다른 주기를 가지는 코사인과 사인의 linear combination으로 표현이 가능합니다. 
- Periodic rectangular wave를 통해 이를 확인해 보는 예제입니다.

```
import numpy as np
from matplotlib import pyplot as plt
```

- 100개의 숫자를 -2pi와 2pi사이에서 생성하여 x라 합니다.

```
x = np.linspace(-2*np.pi,2*np.pi,100)
#print(x)
```

- cos(nx)과 sin(nx)를 함수로 정의합니다.

```
def cos(x,n):
  return np.cos(n*x)
def sin(x,n):
  return np.sin(n*x)
```

- cos(nx) 또는 sin(nx)의 조합이 어떻게 나타나는 지 살펴봅니다.

```
plt.figure(dpi=80)
plt.grid()
plt.plot(x,sin(x,1) + (1/3)*sin(x,3) + (1/5)*sin(x,5) )
plt.show()
```

- n을 1부터 조금씩 늘려가면서 n이 커짐에 따라 어떤 함수로 수렴하는지 확인합니다.

```
n = 1000
fx = np.zeros(np.shape(sin(x,0)))
for n in range(1,n,2):
  fx += (1./n)*sin(x,n)
plt.figure(dpi=80)
plt.grid()
plt.plot(x,fx)
plt.show()
```
