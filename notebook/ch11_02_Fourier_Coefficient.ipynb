```
import numpy as np
from matplotlib import pyplot as plt
```

- 아래와 같은 주기함수의 Fourier coefficient는 a0 = 1/L, an = (2/L)(sin(n *pi* / L) / (n *pi*/L) ) 이고, bn = 0이다. 
- Ch11.7 example 1을 참고

image.png

- N을 충분히 크게하고, L을 바꾸어 가며 coefficient의 변화를 살펴봅니다.

```
L = 16
N = 64
n = np.arange(1,N+1)
An = np.zeros(n.shape)
An = (2/L)*(np.sin(n*np.pi/L))/(n*np.pi/L)
Bn = np.zeros(n.shape)
```

```
plt.figure(dpi=100)

plt.subplot(121)
plt.stem(n,An)
plt.title('An')
plt.xlabel('n')
plt.ylabel('coefficient')
plt.grid('on')
plt.axis([-1,65,min(An),max(An)])

plt.subplot(122)
plt.stem(n,Bn)
plt.title('Bn')
plt.xlabel('n')
plt.ylabel('coefficient')
plt.grid('on')
plt.axis([-1,65,min(An),max(An)])
plt.tight_layout()
plt.show()
```

- 유한한 개수의 Fourier coefficient를 통하여 원래의 함수를 근사했을 때, 사용한 coefficient의 개수에 따라 근사된 결과의 변화를 살펴봅시다.

```
x = np.linspace(-2*L,2*L,1000)
fx = np.zeros_like(x)
a = 4
k = 0
for an in An[:a]:
  k=k+1
  fx += an * np.cos((k*np.pi/L)*x)
plt.plot(x,fx)
plt.axis([-L*1.2,L*1.2,min(fx),max(fx)])
plt.xlabel('x')
plt.ylabel('fx_approximation')
plt.title(f'max.n={a}')
plt.grid('on')
plt.show()
```
