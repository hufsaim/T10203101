```python
import numpy as np
from matplotlib import pyplot as plt
```

## Exact ODE의 implcit solution u(x,y) = c 를 이해하기

- u(x, y)를 정의합니다.

```python
def u(x,y):
  return x**2 + y**2
```

```python
print(u(1,1))
print(u(3,4))
```

- x-y plane에서 u(x, y)를 관찰하기 위해 meshgrid를 이용하겠습니다.

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
xv, yv = np.meshgrid(x,y)
```

```python
plt.imshow(xv,cmap='jet')
plt.axis('scaled')
plt.colorbar()
plt.show()
```

```python
plt.imshow(yv,cmap='jet')
plt.axis('scaled')
plt.colorbar()
plt.show()
```

- 생성한 meshgrid를 통해 미리 정의한 x-y plane의 각 점에서의 u(x,y)를 계산합니다.

```python
uv = u(xv,yv)
```

```python
plt.imshow(uv,cmap='jet')
plt.axis('scaled')
plt.colorbar()
plt.show()
```

- 특별한 c에 대한 contour를 관찰하여 particular solution을 이해합니다.

```python
levels = [1,5,10,20,30]
cs = plt.contour(x,y,uv,levels)
plt.clabel(cs, levels, fontsize=10)
plt.colorbar()
plt.axis('scaled')
plt.grid('on')
```
